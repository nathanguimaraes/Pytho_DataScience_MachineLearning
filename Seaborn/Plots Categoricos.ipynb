# Plots Categoricos
Agora vamos discutir como usar seaborn para tra√ßar dados categoricos. Existem alguns tipos de argumentos principais para isso:

-factorplot(foi substituido nas novas versoes por catplot)

-boxplot

-violinplot

-striplot

-swarmplot

-barplot

-countplot

Vamos passar por exeplos de cada um.

```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips) #o plot nos diz que a media dos valores maxima foi 20
                                                # a media do desvio padrao √© representada pela linha vertical 
                                                # no meio da barra
```

```python
sns.barplot(x='sex', y='total_bill',  data=tips, estimator=np.std) 
#o parametro esstimator nos mostra o valor do desvio padrao  dos desvios padroes 
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette = 'rainbow', hue='sex')
#definindo boxplot por dias 
#pallete define as cores
#hue esta subdividindo os dados e nos dando mais visualizacoes 
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
#plotando dados categoricos na horizontal (orient='h')
```

```python
sns.violinplot(x='day', y='tip', data=tips)
#graficos em forma de violinos, baseado em modelos nao parametricos 
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, )#jitter=true
#split quebra os padroes entre homes e mulheres
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='sex', y='total_bill', data=tips ) 
#utilizando kind=bar temos um brafico em barras 
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar') 
#semelhante ao barplot
```
