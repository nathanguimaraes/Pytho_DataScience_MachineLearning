```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('darkgrid')#ticks, whitegrid, dark 
sns.countplot(x='sex', data=tips)
sns.despine ()#retira bordas da direita e superior
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, height=5, aspect=4) #
#o comando size foi substituido por height 
```

```python
sns.set_context('poster', font_scale=1)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='inferno')
```
