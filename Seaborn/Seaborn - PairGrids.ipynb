```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris['species'].value_counts()
```

```python
#pairgrid permite a criacao de grafico mais poderosa
g = sns.PairGrid(iris)

g.map(plt.scatter)
#cria graficos de dispero utilizando todas as variaveis numeircas possiveis
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)  #aplica uma funcao pra todo o conjunto de dados na diagonal em forma de histograma 
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot) 
```

```python
sns.pairplot(iris, hue='species')
```
