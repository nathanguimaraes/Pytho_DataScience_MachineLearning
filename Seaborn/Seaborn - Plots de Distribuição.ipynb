# Plots de distribuições

vamos discutir alguns gráficos  que nos permite  vizualizar a distribuição de um conjunto de dados. Esses plots são:

-distplot(esse ira mudar nas versoes mais recentes para displot)

-jointpplot

-pairplot

-rugplot

-kdeplot

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
#sns.distplot(tips['total_bill']) #plotando a coluna totalbill
#sns.distplot(tips['total_bill'], kde=False) 
#kde = estimativa de densidade do kernel 
sns.displot(tips['total_bill'], bins=50) #bins divide nosso grafico de barras 
                                        #em menores partes dando uma analise mais detalhado do caso 
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
# jointplot = distibuição conjunta
#kind =  regressão linear
```

```python
sns.pairplot(tips, hue='sex')  # exibe uma matriz com graficos em regressao 
                            #de todos os dados numericos 
                            #hue = adiciona cores diferentes nos graficos
```

```python
sns.rugplot(tips['total_bill']) #util para criar kde(pesquisar conceito)
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])  #fazendo mais sentindo 
```
