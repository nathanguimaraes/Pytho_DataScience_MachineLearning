```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig, ax = plt.subplots()   #essa funcao cria duas instancias a figura e os eixos 
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x * x
```

```python
x
```

```python
y
```

```python
fig, ax=plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
x
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)   #criando figuras
```

```python
ax
```

```python
 for axe in ax:              # plotando graficos ultilizando for
        axe.plot(x, y, 'b')
        axe.set_title('titulo')
fig
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('Titulo')       #outra notacao plotando graficos uma a um

ax[1].plot(x, y, 'b')
ax[1].set_title('Titulo')

fig
```

```python
fig, ax = plt.subplots(nrows = 5, ncols=5)
plt.tight_layout()  #evita a sobreposicao no layoutu
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8, 3))
axes.plot(x, y, 'r-', label='Dados 1')
axes.plot(y, x, 'g--', label='Dados 2')
axes.set_title('Título')
axes.legend(loc=0)   # legendas com base na localizzacao(loc=0)
```
