```python
fig = plt.figure()       #forma orientada a objetos (mais eficaz)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])   #coordenadas do grafico 1
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])   #coordenadas do grafico 2

axes1.plot(x, y)                            #exibindo grafico
axes1.set_xlabel('Eixo X')                  #definindo label
axes1.set_title('Titulo')                   #definindo titulo
axes1.plot(y, x)
axes2.plot(x, y)
```
