```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #comando magico para rederizar graficos 
```

```python
x = np.linspace(0, 5, 50)  #declarando array 
```

```python
len(x)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))   #criando figura figsize define o tamanho
ax.plot(x, x**2, 'g--')    #podemos estar definindo cores e formarto das linhas
ax.plot(x, x**3, 'b-.')    #podemos usar o paramentro color( ax.plot(x, x**2, color='#8800') 
ax.plot(x, x**4, color='#FF008B', linewidth=5, alpha=0.5, linestyle=':')#podemos usar o linewidth para indicar a expessura 
                                                                    
ax.set_xlim([0,2]) #define o limite das linhas no eixo x            #paramentro alpha nos pemite manipular a transparencia 
                                                                    
ax.set_ylim([0, 10])   #define o limite no eixo y                   #paramentro linesyle permite alteramos o estilo        
                                                                                
```
